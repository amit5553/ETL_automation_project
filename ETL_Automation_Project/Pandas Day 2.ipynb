# Filtering (Similar to where clause in SQL)

```python
df =  pd.read_csv("emp.csv",parse_dates=['doj'])
```

```python
df
```

```python
# Get me all the employees who are geting salary more than 3000 
# select * from emp where salary > 3000
salary_300 = df['salary'] > 3000
```

```python
df[salary_300]
```

```python
# Get me all the employees who are geting salary more than 3000 and employees should belong from deptno=20
# select * from emp where salary > 3000 and deptno=20

salary_300 = df['salary'] > 3000
dept_10  = df['deptno'] == 20
final_filter = salary_300 & dept_10
df[final_filter]
```

# Sorting ( arranging data in ascending/descending order)

```python
# Sorting based on eno in ascending order
# select * from emp order by eno
df.sort_values(by='eno')
# or
# select * from emp order by eno asc
df.sort_values(by='eno',ascending=True)
```

```python
# Sorting based on eno in descending order
# select * from emp order by eno desc
df.sort_values(by='eno',ascending=False)
```

```python
# Sorting based on depnto in ascending and  salary in descending order
# select * from emp order by deptno asc,salary desc
df.sort_values(by=['deptno','salary'],ascending=[True,False])
```

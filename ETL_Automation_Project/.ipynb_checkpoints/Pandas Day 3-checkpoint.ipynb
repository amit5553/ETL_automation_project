# Merge ( Similar to SQL joins)

```python
df_emp = pd.read_csv("employees.csv")
df_sal = pd.read_csv("salary.csv")
```

```python
df_sal.head()
```

```python
df_emp.head()
```

```python

# Inner join
# SQL : select e.eno,e.ename,s.sal from employees as e inner join salary as s on e.eno = s.eno
df_emp.merge(df_sal,on='eno',how='inner')
```

```python
# Inner join where common key is different name
# SQL : select e.eno,e.ename,s.sal from employees as e inner join salary as s on e.eid = s.eno
df_emp.merge(df_sal,left_on='eid',right_on='eno',how='inner')
```
